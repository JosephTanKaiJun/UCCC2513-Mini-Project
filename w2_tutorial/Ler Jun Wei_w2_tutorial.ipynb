````python
#2
#function to determine value is odd or even
def isOddOrEven(a):
    if a%2 == 0:
        return True
    else:
        return False

#test case
a = int(input("Enter a number to determine it is odd or even: "))
if isOddOrEven(a):
    print(f"{a} is even")
else:
    print(f"{a} is odd")
````

````python
#3
import numpy as np
import time

# large array size
n = 10000000

#Python List
start_time = time.time()
py_list = list(range(n))
py_result = [i**2 for i in py_list]
py_time = time.time() - start_time

#NumPy array
start_time = time.time()
np_array = np.arange(n)
np_result = np_array**2
np_time = time.time() - start_time

print(f"Python list time: {py_time:.6f} seconds")
print(f"NumPy array time: {np_time:.6f} seconds")
print(f"NumPy is {py_time/np_time:.2f}x faster")

````

````python
#4
# create a 1D array
my_arr = np.arange(10)
print("Initial my_arr: ", my_arr)
arr_slice = np.copy(my_arr) # return carray copy of specified object
print("Initial arr_slice: ", arr_slice)

# change the first element of arr_slice
arr_slice[0] = 55

print("my_arr: ", my_arr)
print("arr_slice: ", arr_slice)
````

````python
#5
import matplotlib.pyplot as plt

black = np.zeros((200,200), dtype="uint8")

x, y = np.ogrid[-100:100, -100:100] #coordinate of white circle
whiteCircle = x*x + y*y <= 30*30 #x^2 + y^2 <= r^2, let r = 30 (radius)
black[whiteCircle] = 1 #all pixel defined is set to 1 (white)

plt.imshow(black, cmap ='gray')
plt.xticks([]), plt.yticks([])
plt.show()
````
